```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
data
```

```python
dataT = np.transpose(data)
dataT
```

```python
y = pd.read_csv('Y.csv')
y.drop('Unnamed: 0',axis=1,inplace=True)
y = y.set_index(dataT.index)
y
```

```python
combined = pd.concat([dataT, y], axis=1)
combined
```

```python
combined = combined.loc[(combined['day'] == 'd0' ) & (combined['cov'] =='Yes') & (combined['icu'] != 'None')]
combined
```

```python
combined = combined.drop(['cov','day'],axis=1)
combined
```

```python
X = combined.drop(['batch','icu'],axis=1)
y= combined['icu']
```

# PCA

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_std)
```

```python
pca.explained_variance_ratio_.sum()
```

```python
ERV = pca.explained_variance_ratio_
ERV
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2', 'PC3'])

pca_df['icu'] = combined['icu'].values

sns.pairplot(pca_df, hue='icu', palette='viridis', diag_kind='kde')
plt.suptitle('Pairplot with ICU as Color Code', y=1.02)
plt.savefig('PCA_ICU.png', bbox_inches='tight')
plt.show()
```

# T-SNE

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.manifold import TSNE

scaler = MinMaxScaler()
X_std = scaler.fit_transform(X)

tsne = TSNE(n_components=3, random_state=42)
X_tsne = tsne.fit_transform(X_std)
import seaborn as sns
import matplotlib.pyplot as plt
tsne_df = pd.DataFrame(data=X_tsne, columns=['C1', 'C2', 'C3'])

tsne_df['icu'] = combined['icu'].values

sns.pairplot(tsne_df, hue='icu', palette='viridis', diag_kind='kde')
plt.suptitle('Pairplot with ICU as Color Code', y=1.02)
plt.savefig('tsne_icu.png',bbox_inches='tight')
plt.show()
```

# Batch Effect

```python
pca_df = pca_df.set_index(combined.index)
pca_df['batch'] = combined['batch']
pca_df
```

```python
pca_df.to_csv('pca_df.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

pca_plot_df = pca_df.copy()

pca_plot_df['icu'] = pca_plot_df['icu'].astype('category')
pca_plot_df['batch'] = pca_plot_df['batch'].astype('category')

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

scatter1 = sns.scatterplot(data=pca_plot_df, x='PC1', y='PC2', hue='batch', style='icu', palette='viridis', alpha=0.7, ax=axes[0])
axes[0].set_xlabel(f'PC1 - Explained Variance: {(pca.explained_variance_ratio_[0] * 100):.2f}%')
axes[0].set_ylabel(f'PC2 - Explained Variance: {(pca.explained_variance_ratio_[1] * 100):.2f}%')
scatter1.legend_.remove()  # Remove legend from the first plot

scatter2 = sns.scatterplot(data=pca_plot_df, x='PC2', y='PC3', hue='batch', style='icu', palette='viridis', alpha=0.7, ax=axes[1])
axes[1].set_xlabel(f'PC2 - Explained Variance: {(pca.explained_variance_ratio_[1] * 100):.2f}%')
axes[1].set_ylabel(f'PC3 - Explained Variance: {(pca.explained_variance_ratio_[2] * 100):.2f}%')
scatter2.legend_.remove()  # Remove legend from the second plot

scatter3 = sns.scatterplot(data=pca_plot_df, x='PC1', y='PC3', hue='batch', style='icu', palette='viridis', alpha=0.7, ax=axes[2])
axes[2].set_xlabel(f'PC1 - Explained Variance: {(pca.explained_variance_ratio_[0] * 100):.2f}%')
axes[2].set_ylabel(f'PC3 - Explained Variance: {(pca.explained_variance_ratio_[2] * 100):.2f}%')
scatter3.legend_.remove()  # Remove legend from the third plot

handles, labels = scatter1.get_legend_handles_labels()
fig.legend(handles, labels, loc='center right')

plt.title('Scatter Plot of the Different Principal Components Grouped By Batch and ICU', loc='center')
plt.savefig('batcheffect.png',bbox_inches='tight')
plt.show()
```

```python
import numpy as np

difference_measure = combined.groupby('batch').agg(np.var)  # Use np.var for variance

sorted_features = difference_measure.mean(axis=0).sort_values(ascending=False)

top_3_features = sorted_features.index[:3]

print("Top 3 features with the most different statistical properties:")
print(top_3_features)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 4, figsize=(12, 4))

batch_groups = combined.groupby('batch')

for i, value in enumerate(['20','501', '328', '1382']):
    data = [group[f'pos.result.{value}'] for name, group in batch_groups]
    ax = axes[i]
    ax.boxplot(data, labels=batch_groups.groups.keys())
    ax.set_title(f'pos.result.{value} by Batches')
    ax.set_xlabel('Batch')
    ax.set_ylabel(f'pos.result.{value}')

plt.tight_layout()
plt.savefig('boxplot.png',bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(pca_df['batch'], pca_df['icu'])
chi2, p_value, _, _ = chi2_contingency(contingency_table)

print(f'Chi-squared statistic: {chi2}')
print(f'P-value: {p_value}')

alpha = 0.05
if p_value < alpha:
    print('There is a statistically significant association between batches and the ICU variable.')
else:
    print('There is no statistically significant association between batches and the ICU variable.')
```

```python
features = ['PC1', 'PC2', 'PC3']
fig, axes = plt.subplots(nrows=1, ncols=len(features), figsize=(15, 6))

for i, feature in enumerate(features):
    sns.violinplot(x='batch', y=feature, data=pca_df, ax=axes[i])
    axes[i].set_title(f'Violin Plot for {feature} by Batch')

plt.tight_layout()
plt.savefig('violin_plot.png',bbox_inches='tight')
plt.show()
```

# Classification


```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
X = combined.drop(['batch','icu'],axis=1)
y= combined['icu']
y = label_encoder.fit_transform(combined['icu'])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.cross_decomposition import PLSRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=32)

rf_classifier = RandomForestClassifier(random_state=42)
rf_cv_predictions = cross_val_predict(rf_classifier, X_train, y_train, cv=5)
rf_classifier.fit(X_train, y_train)
rf_feature_importances = rf_classifier.feature_importances_
rf_top_10_features = X_train.columns[np.argsort(rf_feature_importances)[::-1][:10]]
print("Random Forest Top 10 Features:")
print(rf_top_10_features)
print("Random Forest Classifier Metrics:")
print("Accuracy:", accuracy_score(y_train, rf_cv_predictions))
print("Classification Report:\n", classification_report(y_train, rf_cv_predictions))
print("Confusion Matrix:\n", confusion_matrix(y_train, rf_cv_predictions))

xgb_classifier = XGBClassifier(random_state=42)
xgb_cv_predictions = cross_val_predict(xgb_classifier, X_train, y_train, cv=5)
xgb_classifier.fit(X_train, y_train)
xgb_feature_importances = xgb_classifier.feature_importances_
xgb_top_10_features = X_train.columns[np.argsort(xgb_feature_importances)[::-1][:10]]
print("\nXGBoost Top 10 Features:")
print(xgb_top_10_features)
print("\nXGBoost Classifier Metrics:")
print("Accuracy:", accuracy_score(y_train, xgb_cv_predictions))
print("Classification Report:\n", classification_report(y_train, xgb_cv_predictions))
print("Confusion Matrix:\n", confusion_matrix(y_train, xgb_cv_predictions))

pls = PLSRegression(n_components=2)
pls_cv_predictions = (cross_val_predict(pls, X_train, y_train, cv=5)>0.5).astype('uint8')

print("\nPLS Regression Metrics:")
print("Accuracy:", accuracy_score(y_train, pls_cv_predictions))
print("Classification Report:\n", classification_report(y_train, pls_cv_predictions))
print("Confusion Matrix:\n", confusion_matrix(y_train, pls_cv_predictions))
print('\n---------------------------------------\n')

# Test set 
rf_classifier.fit(X_train, y_train)
y_test_pred_rf = rf_classifier.predict(X_test)
print("\nRandom Forest Test Set Metrics:")
print("Accuracy:", accuracy_score(y_test, y_test_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_test_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred_rf))

xgb_classifier.fit(X_train, y_train)
y_test_pred_xgb = xgb_classifier.predict(X_test)
print("\nXGBoost Test Set Metrics:")
print("Accuracy:", accuracy_score(y_test, y_test_pred_xgb))
print("Classification Report:\n", classification_report(y_test, y_test_pred_xgb))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred_xgb))

pls.fit(X_train, y_train)
y_test_pred_pls = (pls.predict(X_test) > 0.5).astype('uint8')
print("\nPLS Test Set Metrics:")
print("Accuracy:", accuracy_score(y_test, y_test_pred_pls))
print("Classification Report:\n", classification_report(y_test, y_test_pred_pls))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred_pls))
```

```python
print("Random Forest Top 10 Features:")
for feature in rf_top_10_features:
    kegg_ids_with_value_one = list(meta.loc[feature][meta.loc[feature] == 1].index)
    print(f"{feature}: {', '.join(kegg_ids_with_value_one)}")

print("\nXGBoost Top 10 Features:")
for feature in xgb_top_10_features:
    kegg_ids_with_value_one = list(meta.loc[feature][meta.loc[feature] == 1].index)
    print(f"{feature}: {', '.join(kegg_ids_with_value_one)}")
```

```python
common_top_features = set(rf_top_10_features).intersection(xgb_top_10_features)

print("Features in both RF and XGBoost Top 10:")
print(", ".join(common_top_features))
```

```python
rf_kegg_ids_set = set()
xgb_kegg_ids_set = set()
for feature in rf_top_10_features:
    kegg_ids_with_value_one = set(meta.loc[feature][meta.loc[feature] == 1].index)
    rf_kegg_ids_set.update(kegg_ids_with_value_one)
for feature in xgb_top_10_features:
    kegg_ids_with_value_one = set(meta.loc[feature][meta.loc[feature] == 1].index)
    xgb_kegg_ids_set.update(kegg_ids_with_value_one)

common_kegg_ids = rf_kegg_ids_set.intersection(xgb_kegg_ids_set)

print("Common KEGG IDs in RF and XGBoost Top Features:")
print(", ".join(common_kegg_ids))
```
